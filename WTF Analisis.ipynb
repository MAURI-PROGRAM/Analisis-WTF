```python
import pandas as pd
import pandas_datareader.data as web
```

```python
#Conectando con la API de yahoo
wft = web.DataReader("WFT", 'yahoo', '2018-1-1 00:00:00', '2018-9-30 23:59:59')
```

```python
wft.to_csv('base.csv')
```

```python
wft.head(5)
```

```python
#filtro por rango
wft['2018-09-01':'2018-09-20']
```

```python
# Seleccionar todos los campos del registro al 2018-09-20
wft.loc['2018-09-20']
```

```python
#seleccionar el valor de un campo especifico
wft['Adj Close']['2018-09-20']
```

```python
# filtrando por todo un mes
wft['2018-2']
```

```python
# Valores al cierre de cada mes.
wft.asfreq('M', method='ffill')
```

```python
# Valores al cierre de cada mes (días laborales).
wft.asfreq('BM', method='ffill')
```

```python
# valores al cierre de cada trimestre
wft.asfreq('BQ', method='ffill')
```

```python
# desplazando el 1 dia el valor de cierre
desplazado = wft['Adj Close'].shift(1)
```

```python
desplazado.head()
```

```python
# calculando el porcentaje de variación del día.
variacion_diaria = wft['Adj Close'] / wft['Adj Close'].shift(1)-1
wft['var_diaria'] = variacion_diaria
```

```python
wft['var_diaria'].head()
```

```python
# mismo resultado utilizando pct_change()
wft['Adj Close'].pct_change().tail()
```

```python
# calculando rendimiento acumulado diario
rendimiento_diario = (1 + wft['Adj Close'].pct_change()).cumprod()
wft['rend_diario'] = rendimiento_diario
wft['rend_diario'].head()
```

```python
# graficando Adj Close
wft['Adj Close'].plot()
```

```python
# Calculando promedios móviles cada 6 días
wft_ma = wft['Adj Close'].rolling(5).mean()
wft['prod_mov'] = wft_ma
plot = wft[['prod_mov','Adj Close']].plot(figsize=(10, 8), fontsize=12)
```
